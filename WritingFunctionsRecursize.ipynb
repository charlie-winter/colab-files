```python
#Iterative Solution
def csum_it(n):
  total = 0
  while n>0:
    total +=n
    n = n-1
  print(total)
csum_it(5)
```

```python
def csum_rec(n):
  if n ==1:
    return 1
  else:
    return n + csum_rec(n-1)
csum_rec(978)
```

3. Write a function count_strings that accepts two parameters: a list of strings words and another integer length. Return the number of strings in words that are exactly length long.
For example:
Call                                                    Returns
count_strings(["Up", "Down", "Left", "Right"], 4)       2
count_strings(["Up", "Down", "Left", "Right"], 2)       1
count_strings(["Hello", "Goodbye"], 5)                  1  



```python

def count_it(n,a):
  total=0
  for x in n:
    if len(x)==a:
      total += 1
  print(total)
count_it(["Up", "Down", "Left", "Right"], 2)

def count_rec(n,a):
  if not n:
    return 0
  elif len(n[0]) == a:
    return 1 + count_rec(n[1:], a)
  else:
    return count_rec(n[1:],a)
count_rec(["Up", "Down", "Left", "Right"], 2)
```

4. Write a function star_string that accepts an integer parameter n and returns a string of stars (asterisks) 2^n long (i.e., 2 to the nth power).
For example:
Call            Returns      Reason
star_string(0)  "*"          2^0 = 1
star_string(1)  "**"         2^1 = 2
star_string(2)  "****"       2^2 = 4
star_string(3)  "********"   2^3 = 8


```python
def star_int(n):
  x=2**n
  print('*'*x)
star_int(4)
```
